```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n, k = 7, 4
```

```python
logarithmic = [np.log2(i+1) for i in range(n)]
logarithmic
```

```python
linear = [i+1 for i in range(n)]
linear
```

```python
quadratic = [(i+1)**2 for i in range(n)]
quadratic
```

```python
cubic = [(i+1)**3 for i in range(n)]
cubic
```

```python
polynomial = [(i+1)**k for i in range(n)]
polynomial
```

```python
plt.figure(figsize=(9, 9))
plt.plot([(i+1) for i in range(n)], logarithmic) # Θ(log n)
plt.plot([(i+1) for i in range(n)], linear) # Θ(n)
plt.plot([(i+1) for i in range(n)], cubic) # Θ(n^2)
plt.plot([(i+1) for i in range(n)], quadratic) # Θ(n^3)
plt.plot([(i+1) for i in range(n)], polynomial) # Θ(n^k)
plt.legend(['Logarithmic', 'Linear', 'Cubic', 'Quadratic', 'Polynomial'])
plt.xlabel('No of Input')
plt.ylabel('Θ (cost)')

plt.savefig('Comparison.jpg')
plt.show()
```
